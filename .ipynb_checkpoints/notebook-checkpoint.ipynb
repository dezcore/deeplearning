```python
import numpy as np
import matplotlib.pyplot as plt

# Créer des données pour x et y
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Créer un graphique
plt.plot(x, y)
plt.title("Graphique sinusoïde")
plt.xlabel("x")
plt.ylabel("sin(x)")

# Afficher le graphique
plt.show()
```
